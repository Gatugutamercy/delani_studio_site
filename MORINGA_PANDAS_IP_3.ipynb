```
# importing pandas, numpy and matplotlib
import pandas as pd
import numpy as np
import sys
import matplotlib.pyplot as plt
```

```
#Reading the daty 1 excel sheet into the dataframe
telcom1_url = 'http://bit.ly/Telcom_dataset1'
telcom1 = pd.read_csv(telcom1_url)
telcom1
```

```
#we need to change the telecom 1 datafraneme column name from produtc to product and DATETIME to DATE_TIME

telcom1.columns = ['PRODUCT',	'VALUE',	'DATE_TIME',	'CELL_ON_SITE',	'DW_A_NUMBER',	'DW_B_NUMBER',	'COUNTRY_A',	'COUNTRY_B',	'CELL_ID',	'SITE_ID']
telcom1
```

```
#reading daY 2 data
telcom2_url = 'http://bit.ly/Telcom_dataset2'
telcom2 = pd.read_csv(telcom2_url)
telcom2
```

```
# reading day 3 data
telcom3_url = 'http://bit.ly/Telcom_dataset3'
telcom3 = pd.read_csv(telcom3_url)
telcom3
```

```
# we need to change the telecom 1 datafraneme column name from SIET to SITE
telcom3.columns = ['PRODUCT',	'VALUE',	'DATE_TIME',	'CELL_ON_SITE',	'DW_A_NUMBER',	'DW_B_NUMBER',	'COUNTRY_A',	'COUNTRY_B',	'CELL_ID',	'SITE_ID']
telcom3
```

```
# Combining the telecom datasets
telcom_df_combined = pd.concat([telcom1, telcom2, telcom3],  ignore_index=True)
telcom_df_combined
```

```
# dropping for the combined dataset
telcom_df_combined1= telcom_df_combined.drop(['CELL_ON_SITE', 'COUNTRY_A', 'COUNTRY_B'], axis=1)
telcom_df_combined1
```

```
# to drop duplicates
telcom_df_combined2 = telcom_df_combined1.drop_duplicates()
telcom_df_combined2
```

```
#reading the cells_geo data
pd.read_csv('cells_geo.csv')
cells_geo
```

```
cells_geo.columns = ['Unnamed: 0', 'VILLES',	'STATUS',	'LOCALISATION',	'DECOUPZONE',	'ZONENAME',	'LONGITUDE',	'LATITUDE',	'REGION',	'AREA',	'CELL_ID',	'SITE_ID']
cells_geo
```

```
combined_dataset = pd.merge(telcom_df_combined1, cells_geo,
how='outer', on='SITE_ID')
combined_dataset
```
